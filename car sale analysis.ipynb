```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ax = sns.countplot(x = 'Model', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
a = df.groupby(['Model'], as_index=False)['Engine size'].sum().sort_values(by='Engine size', ascending=False)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data = a, x = 'Model',y= 'Engine size')
```

```python
b = df.groupby(['Vehicle type'], as_index=False)['Horsepower'].sum().sort_values(by='Horsepower', ascending=False)
sns.set(rc={'figure.figsize':(50,10)})
sns.barplot(data = b, x = 'Vehicle type',y= 'Horsepower')
```

```python
c = df.groupby(['Model'], as_index=False)['Manufacturer'].sum().sort_values(by='Manufacturer', ascending=False)
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(data = c, x = 'Model',y= 'Manufacturer')
```
